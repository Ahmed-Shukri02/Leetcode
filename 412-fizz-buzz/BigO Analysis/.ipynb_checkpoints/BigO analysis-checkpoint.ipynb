```python
import numpy as np
# first we create a bunch of random integers to use
int_array = np.random.randint(1, 1e4, 1000)
```

```python
# now run the algorithm
def fizzBuzz(n):
        """
        :type n: int
        :rtype: List[str]
        """
        # complexities:
        # Time: O(n)
        # Space: O(n)
        
        final = []
        
        for i in range(n):
            final_text = ""
            if((i+1) % 3 == 0): final_text += "Fizz"
            if ((i+1) % 5 == 0): final_text += "Buzz"
                
            if(final_text == ""): final_text = str(i+1)
            final.append(final_text)
        
        return final
```

```python
import timeit

#calculate runtime of algorithm
def CalculateRuntime(fun: callable, args: tuple) -> float:
    start = timeit.default_timer()
    fun(args)
    stop = timeit.default_timer()
    return (stop-start)
```

```python
import matplotlib.pyplot as plt

# set parameters for plotting and plot time complexity
params = {"text.color" : "gray",
          "axes.labelcolor": "gray",
          "xtick.color" : "gray",
          "ytick.color" : "gray"}

plt.rcParams.update(params)

fig, ax = plt.subplots(figsize = (9, 9))
ax.set_xlabel("Length (n)", fontsize = 15,)
ax.set_ylabel(r"Runtime (ms)", fontsize = 15,)
ax.set_title("Time complexity of algorithm ( O(n) )", fontsize = 25,)

xcol, ycol = [], []
for num in int_array:
    runTime = CalculateRuntime(fizzBuzz, num)
    xcol.append(num)
    ycol.append(runTime * 1000)


ax.set_xlim(0, max(xcol))
ax.plot(xcol, ycol, "rx")
```
